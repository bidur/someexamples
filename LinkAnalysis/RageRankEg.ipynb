# Python code that uses the NetworkX library to estimate the PageRank of a toy dataset:



This code creates a directed graph with the edges specified in the G.add_edges_from() function. Then it uses the nx.pagerank() function to estimate the PageRank for each node in the graph. The resulting PageRank values are stored in the pr dictionary, where the keys are the nodes and the values are the PageRank scores. Finally, the code iterates through the pr dictionary and prints the PageRank of each node.


```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph() #[(0, 3), (1, 3), (2, 4), (3, 5), (3, 6), (4, 6), (5, 6)])

# Add edges to the graph
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'),  ('C', 'D'), ('D', 'B'), ('E', 'D'),('D', 'F'), ('A', 'F')])

# group nodes by column
left_nodes = ['A', 'B', 'C']
middle_nodes = ['D','E']
right_nodes = ['F']

# set the position according to column (x-coord)
pos = {n: (0, i) for i, n in enumerate(left_nodes)}
pos.update({n: (1, i + 0.5) for i, n in enumerate(middle_nodes)})
pos.update({n: (2, i + 0.5) for i, n in enumerate(right_nodes)})

nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
# Compute PageRank
pr = nx.pagerank(G)

# Print the PageRank of each node
for node, rank in pr.items():
    print(f'{node}: {rank:.4f}')
```
